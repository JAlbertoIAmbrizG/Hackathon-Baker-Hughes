## Modelo RandomForest

### Preprocesamiento de datos para Random Forest

Random Forest es un algoritmo que nos permite trabajar tanto con datos categoricos como con datos numericos. Esta gran ventaja al igual que su facil interpretacion y el poco preprocesamiento de datos antes de ocupar, . Lo colocan como un modelo facil de implementar que da buenos resultados en especial con variables numericas continuas como es el caso de nuestro problema

Se excluyen los valores NaN en el Modelo para mejora el Perfomance

```python
df_nonan=df_full[df_full["POWER"].notnull()]
```

En base al HeatMap se eligen las Features o columnas de interes para el modelo, de igual forma se identifica las etiquetas de salida o los valores a predecir

```python
#Variables de Entrada
input_cols=['CMP_SPEED', 'CDP', 'GGDP', 'HPT_IT', 'CDT', 'LPT_IT', 'EXH_T']
#Variable de Salida
target_col=['POWER']
```

```python
#Se crea una copia de las columnas
inputs = df_nonan[input_cols].copy()
targets = df_nonan[target_col].copy()
```

Se ralizo una normalizacion de los datos, esto debido a que las regresion son muy sensibles a la escala de los datos. Para esto se eligio usar la siguiente normalizacion

\begin{align}
        X_{STD} = \frac{X-X_{min}}{X_{max}-X_{min}}
            \end{align}

\begin{align}
X{scaled}=X_{STD}(max-min)+min
 \end{align}


```python
#se utilizo la libreria de sklearn para preprocesamiento de datos 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(inputs[input_cols])
inputs[input_cols] = scaler.transform(inputs[input_cols])
```

Se Divide el Dataset para que el modelo pueda ser entrenado y validado para esto se eligio una division de 25% test de validacion y un 75% para el entrenar el modelo

```python
#Division del Data set en en un conjunto de entrenamiento y otro de Validacion
from sklearn.model_selection import train_test_split
train_inputs, val_inputs, train_targets, val_targets = train_test_split(
    inputs[input_cols], targets, test_size=0.25, random_state=42)
```

```python
#importamos el modelo de Randomforest de la libreria de sklearn
from sklearn.ensemble import RandomForestRegressor
```

```python
# se realiza un caso base para poder evaluar los resultados preliminares del modelo
rf1 = RandomForestRegressor(n_jobs=-1, random_state=42)
```

```python
#Entrenamiento del Modelo
rf1.fit(train_inputs, train_targets.values.ravel())
```

```python
#Prediccion del modelo en los datos de entrenamiento
rf1_train_preds = rf1.predict(train_inputs)
```

Para poder evaluar la eficacia del modelo, se calcula El error cuadrático medio (RMSE) mide la cantidad de error que hay entre dos conjuntos de datos

```python
from sklearn.metrics import mean_squared_error
rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
rf1_val_preds = rf1.predict(val_inputs)
rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
print('Train RMSE: {}, Validation RMSE: {}'.format(rf1_train_rmse, rf1_val_rmse))
```

En una primera aproximacion obtuvimos resultados muy buenos lo cual nos motiva a continuar con este modelo. A continuacion podemos graficar y ver la impoartancia que considero el modelo para las diferentes features o columnas utilizadas como input.

```python
importance_df = pd.DataFrame({
    'feature': train_inputs.columns,
    'importance': rf1.feature_importances_
}).sort_values('importance', ascending=False)
```

```python
importance_df
```

```python
plt.title('Feature Importance')
sns.barplot(data=importance_df.head(10), x='importance', y='feature');
```

### Hyperparameter Tuning and Overfitting

Para mejorar el performance de nuestro modelo de RandomForest, utilizamos los parametros del modelo para de esta forma conseguir mejores predicciones. 

El primer parametro a calificar es el numero de arboles considerados en el model. El valor por default que se utiliza son 100. Sin embargo, podemos ver que no es el mejor

```python
def max_depth_error(n_arboles):
  rf1 = RandomForestRegressor(n_jobs=-1, random_state=42,n_estimators=n_arboles)
  rf1.fit(train_inputs, train_targets.values.ravel())
  rf1_train_preds = rf1.predict(train_inputs)
  rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
  rf1_val_preds = rf1.predict(val_inputs)
  rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
  return {'N° Arboles': n_arboles, 'Training Error': rf1_train_rmse, 'Validation Error': rf1_val_rmse}
#print('Train RMSE: {}, Validation RMSE: {}'.format(rf1_train_rmse, rf1_val_rmse))
  

    #model = DecisionTreeClassifier(max_depth=md, random_state=42)
    #model.fit(X_train, train_targets)
    #train_acc = 1 - model.score(X_train, train_targets)
    #val_acc = 1 - model.score(X_val, val_targets)
    #return {'Max Depth': md, 'Training Error': train_acc, 'Validation Error': val_acc}
```

El siguiente parametro a evaluar es max_depth, el cual nos ayuda a prevenir que los arboles de descision caigan en un overfiting, memorizando los ejemplos de entrenamiento. 

```python
def max_depth_error(max_dep):
  rf1 = RandomForestRegressor(n_jobs=-1, random_state=42,max_depth=max_dep)
  rf1.fit(train_inputs, train_targets.values.ravel())
  rf1_train_preds = rf1.predict(train_inputs)
  rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
  rf1_val_preds = rf1.predict(val_inputs)
  rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
  return {'profundidad': max_dep, 'Training Error': rf1_train_rmse, 'Validation Error': rf1_val_rmse}
```

```python
errors_df_depth = pd.DataFrame([max_depth_error(md) for md in range(5, 30,2)])
```

El siguiente parametro a analizar es el numero de caracteristicas o features que se deben de considerar en cada arbol de decision 

```python
def max_depth_error(features):
  rf1 = RandomForestRegressor(n_jobs=-1, random_state=42,max_features=features)
  rf1.fit(train_inputs, train_targets.values.ravel())
  rf1_train_preds = rf1.predict(train_inputs)
  rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
  rf1_val_preds = rf1.predict(val_inputs)
  rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
  return {'Caracteristicas': features, 'Training Error': rf1_train_rmse, 'Validation Error': rf1_val_rmse}
```

```python
errors_df_features = pd.DataFrame([max_depth_error(md) for md in ['auto','log2',1,2,3,4,5,6,7]])
```

```python
errors_df_features # 6 parece ser un buen valor a considerar para cada split
```

Se evaluo el numero minimo de muestras a tomar para cada arbol y el numero de hojas o nodos a considerar para cada uno. Sin embargo al no ser tantas las features consideradas, no hubo cambio en los valores de RMSE

```python
def max_depth_error(a,b):
  rf1 = RandomForestRegressor(n_jobs=-1, random_state=42,min_samples_split=a,min_samples_leaf=b)
  rf1.fit(train_inputs, train_targets.values.ravel())
  rf1_train_preds = rf1.predict(train_inputs)
  rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
  rf1_val_preds = rf1.predict(val_inputs)
  rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
  return {'split': a ,'left':b,'Training Error': rf1_train_rmse, 'Validation Error': rf1_val_rmse}
```

```python
max_depth_error(3,2)
```

```python
max_depth_error(2,2)
```

Por ultimo se evaluo el limite a considerar para las divisiones de los nodos en cada arbol, modificando el valor de indice gini, el valor por default es cero, por lo cual podemos jugar con valores pequeños cercanos a cero

```python
def max_depth_error(gin):
  rf1 = RandomForestRegressor(n_jobs=-1, random_state=42,min_impurity_decrease=gin)
  rf1.fit(train_inputs, train_targets.values.ravel())
  rf1_train_preds = rf1.predict(train_inputs)
  rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
  rf1_val_preds = rf1.predict(val_inputs)
  rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
  return {'gini': gin, 'Training Error': rf1_train_rmse, 'Validation Error': rf1_val_rmse}
```

```python
errors_df_gini = pd.DataFrame([max_depth_error(md) for md in [1e-9,1e-7,1e-5,1e-2]])
```

```python
errors_df_gini # parace producir un cambio 1e-2
```

```python
# Evaluacion del modelo
from sklearn.metrics import mean_squared_error
rf1_train_rmse =  mean_squared_error(train_targets, rf1_train_preds, squared=False)
rf1_val_preds = rf1.predict(val_inputs)
rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds,squared=False)
print('Train RMSE: {}, Validation RMSE: {}'.format(rf1_train_rmse, rf1_val_rmse))
```
